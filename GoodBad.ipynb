```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path

# learning rate
df = pd.read_csv("processed_data\participant_summary.csv")
Learner = df[["Participant_ID","Learner_Group"]]

# distance to target mean
mean_dist = {}

for file in Path(".").glob("*_SESSION2.csv"):
    dfs = pd.read_csv(file)
    participant = file.stem.split("_")[0]
    distance = dfs['distance_to_target'].mean()
    mean_dist[participant] = distance

DataFrame = pd.DataFrame(list(mean_dist.items()),columns=["Participant_ID","distance_to_target"])
complete = pd.merge(Learner, DataFrame, on="Participant_ID")
print(complete)

# Take the mean with std of the distance to target, error bars(add Good learners(n=15), Bad Learners(n=5) )
```
