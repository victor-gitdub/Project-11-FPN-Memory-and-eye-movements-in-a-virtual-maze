```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
# distance to target mean
mean_dist = {}
mean_fix = {}

for file in Path(".").glob("*_SESSION2.csv"):
    dfs = pd.read_csv(file)
    participant = file.stem.split("_")[0]
    distance = dfs['distance_to_target'].mean()
    fixation = dfs['mean_fixation_time'].mean()
    mean_dist[participant] = distance
    mean_fix[participant] = fixation

DataFrame = pd.DataFrame(list(mean_dist.items()),columns=["Participant_ID","distance_to_target"])
DataFram = pd.DataFrame(list(mean_fix.items()),columns=["Participant_ID","mean_fixation_time"])

complete = pd.merge(DataFram, DataFrame, on="Participant_ID")
print(complete)
```
